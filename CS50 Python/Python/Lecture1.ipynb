```python
name = input("Type your full name")

# Remove all the whitespaces
name = name.strip()
# capitalize() - make the first letter of each word capitalized
name = name.capitalize()
# title() - make the first letter of each word capitalized and the rest lowercase
name = name.title()

# In python, you can chain all the functions together
name = name.strip().title()

# or this
name = input("Type your full name").strip().title()

print(f"hello, {name}")
```

## int

```python
x = input("type x") 
y = input("type y")

print(x + y)

z = int(x) + int(y)
print(z)
```

```python
x = float(input("type x")) 
y = float(input("type y"))

print(x + y)

z = round(x+y)
t = round(x / y, 2)
print(f"{z:,}")
print(f"{z:.2f}")
print(t)
```
