```python
while True:
    n = int(input("Enter the numebr"))
    if n > 3:
        break
```

```python
names = ["giwon", "shwan", "jamal"]
for name in names:
    print(name)

for i in range(len(names)):
    print(names[i])
```

```python
# Dict

names = {
    "Lee" : "Giwon",
    "Kim" : "kifaef",
    "Gorman" : "Shawn"
}

# print(names["Lee"])

for name in names:
    print(name, names[name], sep=",")
```

```python
students = [
    {"name": "giwon", "house": "oihafe"},
    {"name": "asesf", "house": "fasf"},
    {"name": "fefefef", "house": None }
]

for i in range(len(students)):
    print(students[i]["name"] ,students[i]["house"], sep=", ")

for student in students:
    print(student["name"], student["house"])

```

```python
def square(n):
    for i in range(n):
        print("*",end="")


def main(input):
    
    for i in range(input):
        print("$",end="")
        square(n)


a = int(input("Enter the numeber of row"))
main(a)
```

```python
def square(n):
    for _ in range(n):
        print("*", end="")


def main(size):
    for _ in range(size):
        print("$", end="")
        square(size)
        print()  # move to next line


a = int(input("Enter the number of rows: "))
main(a)
```
