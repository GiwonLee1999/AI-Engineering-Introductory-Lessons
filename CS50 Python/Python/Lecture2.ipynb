## Conditionals

```python
x = int(input("Enter x"))
y = int(input("Enter y"))

if x > y:
    print("X is greater than y")
if x < y:
    print("Y is greater than x")
if x == y:
    print("x is equal to y")
```

```python
x = int(input("Enter x"))
y = int(input("Enter y"))

if x > y:
    print(f"X is greater than y. x:{x} y:{y}")
elif x < y:
    print(f"Y is greater than x. x:{x} y:{y}")
else:
    print(f"x is equal to y, x:{x} y:{y}")
```

```python
x = int(input("Enter x"))
y = int(input("Enter y"))

if x > y or x < y:
    print(f"x is not equal to y. x:{x} y:{y}")
elif x <= y and x >= y:
    print(f"x is not equal to y. x:{x} y:{y}")
```

```python
def main():
    x = int(input("Enter x"))
    if is_even(x):
        print("Even")
    else:
        print("odd")

def is_even(x):
    return True if x % 2 == 0 else False

main() 
```

```python
math = input("Expression: ")

x,op,y = math.split(" ")

x = float(x)
y = float(y)

if op == "-":
    print(x-y)
elif op == "+":
    print(x+y)
elif op == "*":
    print(x*y)
elif op == "/":
    print(round(x/y, 1))
```
